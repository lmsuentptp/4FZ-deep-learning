```python
import pdfplumber
import json

def extract_pdf_local(file_path):
    """
    Extracts text from a PDF file located on the system
    """
    text = ''
    
    # Open and extract text from PDF file
    with pdfplumber.open(file_path) as pdf:
        for page in pdf.pages:
            text += page.extract_text()
    
    return text

def save_to_json(company, year, ticker, file_path, save_path):
    """
    Extract content from local PDF and save to a JSON file
    """
    # Extract PDF content
    report_content = extract_pdf_local(file_path)
    
    # Create report dictionary
    report = {'company': company, 'year': year, 'ticker': ticker, 'file_path': file_path, 'content': report_content}
    
    # Save extracted data to JSON file
    json_file_path = save_path + company + year + '.json'
    with open(json_file_path, 'w') as outfile:
        json.dump(report, outfile)
        
    print(f"Report saved to {json_file_path}")

# Example usage: extracting Tesla reports and saving as JSON
file_path = r"C:\Users\HHGiang\Documents\KEIO\4FZ\2023-tesla-impact-report.pdf"  # Path to your PDF file
save_path = r"C:\Users\HHGiang\Documents\KEIO\4FZ\\"  # Path where JSON file will be saved

save_to_json('Tesla', '2023', 'TSLA', file_path, save_path)
```

```python
def remove_non_ascii(text):
    printable = set(string.printable)
    return ''.join(filter(lambda x: x in printable, text))

def not_header(line):
    # as we're consolidating broken lines into paragraphs, we want to make sure not to include headers
    return not line.isupper()

def extract_sentences(nlp, text):
    """
    Extracting ESG statements from raw text by removing junk, URLs, etc.
    We group consecutive lines into paragraphs and use spacy to parse sentences.
    """
    MIN_WORDS_PER_PAGE = 500
    
    pages = text.split('##PAGE_BREAK##')
#     print('Number of Pages: {}'.format(len(pages)))

    lines = []
    for page in pages:
        
        # remove non ASCII characters
        text = remove_non_ascii(page)
        
        if len(text.split(' ')) < MIN_WORDS_PER_PAGE:
#             print('Skipped Page: {}'.format(len(text.split(' '))))
            continue
        
        prev = ""
        for line in text.split('\n\n'):
            # aggregate consecutive lines where text may be broken down
            # only if next line starts with a space or previous does not end with dot.
            if(line.startswith(' ') or not prev.endswith('.')):
                prev = prev + ' ' + line
            else:
                # new paragraph
                lines.append(prev)
                prev = line

        # don't forget left-over paragraph
        lines.append(prev)
        lines.append('##SAME_PAGE##')
        
    lines = '  '.join(lines).split('##SAME_PAGE##')
    
    # clean paragraphs from extra space, unwanted characters, urls, etc.
    # best effort clean up, consider a more versatile cleaner
    
    sentences = []
    pages_content = []

    for line in lines[:-1]:
        # removing header number
        line = re.sub(r'^\s?\d+(.*)$', r'\1', line)
        # removing trailing spaces
        line = line.strip()
        # words may be split between lines, ensure we link them back together
        line = re.sub(r'\s?-\s?', '-', line)
        # remove space prior to punctuation
        line = re.sub(r'\s?([,:;\.])', r'\1', line)
        # ESG contains a lot of figures that are not relevant to grammatical structure
        line = re.sub(r'\d{5,}', r' ', line)
        # remove emails
        line = re.sub(r'\S*@\S*\s?', '', line)
        # remove mentions of URLs
        line = re.sub(r'((http|https)\:\/\/)?[a-zA-Z0-9\.\/\?\:@\-_=#]+\.([a-zA-Z]){2,6}([a-zA-Z0-9\.\&\/\?\:@\-_=#])*', r' ', line)
        # remove multiple spaces
        line = re.sub(r'\s+', ' ', line)
        # join next line with space
        line = re.sub(r' \n', ' ', line)
        line = re.sub(r'.\n', '. ', line)
        line = re.sub(r'\x0c', ' ', line)
        
        pages_content.append(str(line).strip())

        # split paragraphs into well defined sentences using spacy
        for part in list(nlp(line).sents):
            sentences.append(str(part).strip())

#           sentences += nltk.sent_tokenize(line)
            
    # Only interested in full sentences and sentences with 10 to 100 words.
    sentences = [s for s in sentences if re.match('^[A-Z][^?!.]*[?.!]$', s) is not None]
    sentences = [s.replace('\n', ' ') for s in sentences]
    sentences = [s for s in sentences if (len(s.split(' ')) > 10) & (len(s.split(' ')) < 100)]

    return pages_content, sentences
```

```python
def run_NLP(content):

    def sent_to_words(sentences):
        for sentence in sentences:
            yield(gensim.utils.simple_preprocess(str(sentence), deacc=True))  # deacc=True removes punctuations

    # Define functions for stopwords, bigrams, trigrams and lemmatization
    def remove_stopwords(texts):
        return [[word for word in simple_preprocess(str(doc)) if word not in stop_words] for doc in texts]

    def make_bigrams(texts):
        return [bigram_mod[doc] for doc in texts]

    def make_trigrams(texts):
        return [trigram_mod[bigram_mod[doc]] for doc in texts]

    def lemmatization(texts, allowed_postags=['NOUN', 'ADJ', 'VERB', 'ADV']):
        """https://spacy.io/api/annotation"""
        texts_out = []
        for sent in texts:
            doc = nlp(" ".join(sent)) 
            texts_out.append([token.lemma_ for token in doc if token.pos_ in allowed_postags])
        return texts_out

    data_words = list(sent_to_words(content))

    # Build the bigram and trigram models
    bigram = gensim.models.Phrases(data_words, min_count=5, threshold=100) # higher threshold fewer phrases.
    trigram = gensim.models.Phrases(bigram[data_words], threshold=100)  

    # Faster way to get a sentence clubbed as a trigram/bigram
    bigram_mod = gensim.models.phrases.Phraser(bigram)
    trigram_mod = gensim.models.phrases.Phraser(trigram)

    # Remove Stop Words
    data_words_nostops = remove_stopwords(data_words)

    # Form Bigrams
    data_words_bigrams = make_bigrams(data_words_nostops)

    # Do lemmatization keeping only noun, adj, vb, adv
    data_lemmatized = lemmatization(data_words_bigrams, allowed_postags=['NOUN', 'ADJ', 'VERB', 'ADV'])
    
    return data_lemmatized
```

```python
# Create Dictionary
id2word = corpora.Dictionary(data_lemmatized)

# Create Corpus
texts = data_lemmatized

# Term Document Frequency
corpus = [id2word.doc2bow(text) for text in texts]
```

```python
results = []
optimal_score = 0
optimal_model = None

for num_of_topics in tqdm(range(2,15,1)):
    
    result = {}
    
    # Build LDA model
    lda_model = gensim.models.ldamodel.LdaModel(corpus=corpus,
                                                id2word=id2word,
                                                num_topics=num_of_topics, 
                                                random_state=42,
                                                update_every=1,
                                                chunksize=5,
                                                passes=20,
                                                alpha='auto',
                                                per_word_topics=True) 

    coherence_model_lda = CoherenceModel(model=lda_model, texts=data_lemmatized, dictionary=id2word, coherence='c_v')
    
    if optimal_score < coherence_model_lda.get_coherence():
        optimal_score = coherence_model_lda.get_coherence()
        optimal_model = lda_model

    result['num_of_topics'] = num_of_topics
    result['perplexity'] = lda_model.log_perplexity(corpus)
    result['coherence_score'] = coherence_model_lda.get_coherence()
    
    results.append(result)
    
results = pd.DataFrame(results)  
results.set_index('num_of_topics', inplace=True) 
```

```python
file_paths = {
    '2019': 'C:/Users/HHGiang/Documents/KEIO/4FZ/2019-tesla-impact-report.pdf',
    '2020': 'C:/Users/HHGiang/Documents/KEIO/4FZ/2020-tesla-impact-report.pdf',
    '2021': 'C:/Users/HHGiang/Documents/KEIO/4FZ/2021-tesla-impact-report.pdf',
    '2022': 'C:/Users/HHGiang/Documents/KEIO/4FZ/2022-tesla-impact-report.pdf',
    '2023': 'C:/Users/HHGiang/Documents/KEIO/4FZ/2023-tesla-impact-report.pdf'
}

# Initialize the summarization model (T5) and sentiment analysis (BERT)
t5_tokenizer = T5Tokenizer.from_pretrained("t5-base")
t5_model = T5ForConditionalGeneration.from_pretrained("t5-base")

# Summarization function
def summarize_text(text):
    inputs = t5_tokenizer.encode("summarize: " + text, return_tensors="pt", max_length=512, truncation=True)
    summary_ids = t5_model.generate(inputs, max_length=150, min_length=40, length_penalty=2.0, num_beams=4, early_stopping=True)
    return t5_tokenizer.decode(summary_ids[0], skip_special_tokens=True)
```

```python
# ESG-BERT Model Initialization
esg_bert_tokenizer = AutoTokenizer.from_pretrained("nbroad/ESG-BERT")
esg_bert_model = AutoModelForSequenceClassification.from_pretrained("nbroad/ESG-BERT")

# ESG classification categories
categories = [
    "Business Ethics", "Data Security", "Access And Affordability", "Business Model Resilience", 
    "Competitive Behavior", "Critical Incident Risk Management", "Customer Welfare", 
    "Director Removal", "Employee Engagement Inclusion And Diversity", "Employee Health And Safety", 
    "Human Rights And Community Relations", "Labor Practices", "Management Of Legal And Regulatory Framework", 
    "Physical Impacts Of Climate Change", "Product Quality And Safety", "Product Design And Lifecycle Management", 
    "Selling Practices And Product Labeling", "Supply Chain Management", "Systemic Risk Management", 
    "Waste And Hazardous Materials Management", "Water And Wastewater Management", 
    "Air Quality", "Customer Privacy", "Ecological Impacts", "Energy Management", "GHG Emissions"
]
```

```python
# Function for ESG-BERT classification
def classify_esg_bert(text):
    inputs = esg_bert_tokenizer(text, return_tensors="pt", truncation=True, padding=True, max_length=512)
    outputs = esg_bert_model(**inputs)
    logits = outputs.logits.detach().cpu().numpy()
    return {category: score for category, score in zip(categories, logits[0])}

# Extract content from a report file
def extract_report_content(file_path):
    with open(file_path, 'rb') as file:
        # This assumes you have a pdf to text function here.
        return extract_pdf(file)  # Custom function to extract text from PDF

# Example function to run the analysis for each year
def run_bert_analysis_on_reports(file_paths):
    for year, file_path in file_paths.items():
        print(f"Analyzing {year} report...")
        
        # Extract content
        report_content = extract_report_content(file_path)

        # Summarize the report content (optional)
        summary = summarize_text(report_content)
        print(f"Summary for {year}:\n{summary}")
        
        # Classify with ESG BERT
        esg_classification = classify_esg_bert(summary)
        print(f"ESG Classification for {year}:\n{esg_classification}")

# Run the full analysis
run_bert_analysis_on_reports(file_paths)
```

```python
import yfinance as yf

def get_tesla_financial_data(year):
    ticker = yf.Ticker("TSLA")
    
    # Pull financial data for the given year (2019-2023)
    start_date = f'{year}-01-01'
    end_date = f'{year}-12-31'
    stock_data = ticker.history(start=start_date, end=end_date)
    
    # Get market capitalization for the last date of the year
    market_cap = stock_data['Close'][-1] * ticker.info['sharesOutstanding']
    
    # Get total liabilities and total assets from Tesla's financials
    balance_sheet = ticker.balance_sheet
    
    total_liabilities = balance_sheet.loc['Total Liabilities Net Minority Interest'][0]  # Adjust index as needed
    total_assets = balance_sheet.loc['Total Assets'][0]  # Adjust index as needed
    
    # Calculate Tobin's Q
    tobins_q = (market_cap + total_liabilities) / total_assets
    
    return tobins_q, market_cap, total_liabilities, total_assets

# Create a dataframe to store Tobin's Q for each year
tobin_q_data = []

for year in range(2015, 2023 + 1):
    tobins_q, market_cap, liabilities, assets = get_tesla_financial_data(year)
    tobin_q_data.append({'Year': year, 'Tobin Q': tobins_q, 'Market Cap': market_cap, 'Liabilities': liabilities, 'Assets': assets})

df_tobin_q = pd.DataFrame(tobin_q_data)

# Display Tobin's Q DataFrame
print(df_tobin_q)
# Keep only the 'Year' and 'Tobin Q' columns
df_tobin_q = df_tobin_q[['Year', 'Tobin Q']]

# Print the resulting DataFrame
print(df_tobin_q)
```

```python
import statsmodels.api as sm

# Define the dependent and independent variables
X = df_cleaned[['Shareholder Rights', 'Ethics & Compliance', 'Social Pillar Score']]
y = df_cleaned['Tobin Q']

# Add a constant to the model (intercept)
X = sm.add_constant(X)

# Fit the OLS model
model = sm.OLS(y, X).fit()

# Print the regression results
print(model.summary())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm
from scipy import stats

# Get the residuals from the model
residuals = model.resid

# Plotting the Q-Q plot for residuals
sm.qqplot(residuals, line='45')
plt.title('Q-Q Plot of Residuals')
plt.show()

# Jarque-Bera test for normality
jb_test = stats.jarque_bera(residuals)
print('Jarque-Bera Test:', jb_test)
```

```python
# The p-value of the Jarque-Bera test is 0.9177, which is very high, meaning we fail to reject the null hypothesis. This indicates that the residuals are normally distributed, so we are good in terms of normality.
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor

# VIF DataFrame
vif_data = pd.DataFrame()
vif_data["Feature"] = X.columns
vif_data["VIF"] = [variance_inflation_factor(X.values, i) for i in range(len(X.columns))]

print(vif_data)
```

```python
#Ethics & Compliance has  a VIF of 11.26, which is above the critical threshold of 10. This indicates a strong multicollinearity issue, which may distort the estimates of the coefficients and inflate the standard errors.
#A VIF of this magnitude suggests you might want to remove or combine this variable with another related metric or look for highly correlated variables in your regression model.
```

```python
#The Breusch-Pagan test returned a p-value of 0.117, meaning we fail to reject the null hypothesis, so we do not have strong evidence of heteroscedasticity. However, the F-statistic p-value is low, which indicates that there could still be some mild variance inconsistency.
```

```python
# Durbin-Watson test for autocorrelation
durbin_watson_stat = sm.stats.durbin_watson(residuals)
print('Durbin-Watson Statistic:', durbin_watson_stat)
```
